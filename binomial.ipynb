# Binomial Distribution 

```python
import numpy as np 
import matplotlib.pylab as plt
import scipy.stats as s
import seaborn as sns
```

Probability of one coin results one tail

```python

coin_tosses = np.random.binomial(n=1,p=0.5,size=100000000)
print(coin_tosses)
```

```python
frequency = np.count_nonzero(coin_tosses == 1)
print(frequency)
```

```python
relative_frequency = frequency/coin_tosses.shape[0]
print(relative_frequency)
```

Probability of five coin results one tail or frequency of getting one tail while tosing five coins ten times 


```python
coin_tosses = np.random.binomial(n=5,p=0.5,size=100000000)
print(coin_tosses)
```

```python
frequency = np.count_nonzero(coin_tosses == 1)
print(frequency)
```

```python
relative_frequency = frequency/coin_tosses.shape[0]
print(relative_frequency)
```
